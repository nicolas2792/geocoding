```python

def location_cust(customer):
    from geopy.extra.rate_limiter import RateLimiter
    from geopy.geocoders import Nominatim
    if customer.shape[0] ==1:
        direction=customer["address"].loc[0]
        geo = RateLimiter(Nominatim(user_agent="myApp").geocode,min_delay_seconds=1)
        customer["latitude"]=geo(str(direction)).latitude
        customer["longitude"]=geo(str(direction)).longitude
        for i in customer["borrower_needs"]:
            if i <= 1000000:
                customer["asset"]="0-1M"
            elif 1000000 < i <=10000000:
                customer["asset"]="1-10M"
            elif 10000000 < i <=25000000:
                customer["asset"]="10-25M"
            elif  i > 25000000:
                customer["asset"]="25M+"
    else:
        long=[]
        lat=[]
        geo = RateLimiter(Nominatim(user_agent="myApp").geocode,min_delay_seconds=1)
        for i in customer["address"]:
            try:
                long.append(geo(str(i)).longitude)
                lat.append(geo(str(i)).latitude)
            except:
                long.append("Not Found")
                lat.append("Not Found")
        customer["latitude"]=lat
        customer["longitude"]=long  
        for i in customer["borrower_needs"]:
            if i <= 1000000:
                customer["asset"]="0-1M"
            elif 1000000 < i <=10000000:
                customer["asset"]="1-10M"
            elif 10000000 < i <=25000000:
                customer["asset"]="10-25M"
            elif  i > 25000000:
                customer["asset"]="25M+"
    return(customer)
```

```python
def distance(customer):
    from geopy import distance
    if customer.shape[0] ==1:
        query=banks_list[(banks_list["Financial Institution State"]==customer["state"].loc[0]) & 
                         (banks_list["asset"]== customer["asset"].loc[0]) ]
        dist_list=[]
        for a,b in zip(query["latitude"],query["longitude"]):
            try:
                dist_list.append(distance.distance((a,b),(customer["latitude"].loc[0],customer["longitude"].loc[0])).miles)
            except: dist_list.append(10000)
        query["distance"]=dist_list
        query["CLIENT_NAME"]=str(customer["name"].loc[0])
        query=query.sort_values(by="distance",ascending=True).head(5)
    else:
        final_query=[]
        for i,y,z in zip(customer["state"],customer["name"],customer["asset"]):
            query=banks_list[(banks_list["Financial Institution State"]==i) & (banks_list["asset"]==z)]
            cust=customer[customer["name"]==y]
            dist_list=[]
            for a,b in zip(query["latitude"],query["longitude"]):
                try:
                    dist_list.append(distance.distance((a,b),(cust["latitude"].values[0],cust["longitude"].values[0])).miles)
                except: dist_list.append(10000)
            query["distance"]=dist_list
            query["CLIENT_NAME"]=y
            query=query.sort_values(by="distance",ascending=True).head(5)
            final_query.append(query)
        query= pd.concat(final_query, ignore_index=False)  
    return(query)
        
            
```
